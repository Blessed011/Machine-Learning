```python
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
wine = load_wine()
data = pd.DataFrame(data= np.c_[wine['data'], wine['target']],
                     columns= wine['feature_names'] + ['target'])
```

```python
headtail = pd.concat([data.head(5), data.tail(5)])
headtail
```

```python
feature = wine.data[:, 0]   # возьмём первый признак
scaler = StandardScaler()
scaled_feature = scaler.fit_transform(feature.reshape(-1, 1))   # отмасштабируем его
```

```python
scaled_feature
```

```python
target = wine.target   # выберем в качестве категорильного признака целевой
```

```python
label_encoder = LabelEncoder()   # преобразуем категориальные данные в количественные с помощью LabelEncoder
le_target = label_encoder.fit_transform(target)
le_target
```

```python
onehot_encoder = OneHotEncoder(sparse=False)   # преобразуем категориальные данные в количественные с помощью OneHotEncoder
oe_target = onehot_encoder.fit_transform(target.reshape(-1, 1))
oe_target
```

**StandardScaler** используется для масштабирования данных, так как он является распространенным методом для нормализации данных и приведения их к единой шкале.

**Label Encoding** используется для преобразования категориальных признаков в целые числа. Это простой и эффективный метод, который сохраняет порядок категорий и позволяет использовать категориальные признаки в числовых вычислениях.

**One-Hot Encoding** используется для преобразования категориальных признаков в бинарные векторы. Этот метод создает новую колонку для каждой категории, что приводит к увеличению размерности данных, но позволяет использовать категориальные признаки в более сложных моделях машинного обучения, таких как сети глубокого обучения.
